```python
from sklearn.preprocessing import StandardScaler

x=data.drop(['Value', 'User ID', 'Time_Stamp'],axis=1)
y=data["Value"]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=41)


x_scaled_train=StandardScaler().fit_transfrom(x_train)
x_scaled_test=StandardScaler().transform(x_train)
```

```python
from sklearn.linear_model import BayesianRidge
from sklearn.metrics import mean_squared_error
liste=[]
liste1=[]
model=BayesianRidge().fit(x_scaled_train, y_train)
y_pred=model.predict(x_scaled_test)

liste.append(mean_squared_error(y_test,y_pred))
liste1.append(np.sqrt(mean_squared_error(y_test,y_pred)))


print("MSE:",round(np.mean(liste),2))
print("RMSE:",round(np.mean(liste1),2))
```
